```python
OL=OL_2019.sort_values('Overall', ascending=False).groupby('Team').head(5).sort_values(['Position', 'Overall'], ascending=False)
OL_2019 = OL.groupby('Team')['Overall'].describe()

OL_2019 = OL_2019.reset_index()
OL_2019 = OL_2019.set_index('Team')
OL_2019 = OL_2019.reset_index()
OL_2019
```

```python
OL_2019['OLrank'] = OL_2019['mean'].rank(method='first', ascending=False)
OL_2019 = OL_2019[['Team','mean','OLrank']]

OL_2019 = OL_2019.sort_values(by='OLrank')
OL_2019 = OL_2019.reset_index()
OL_2019 = OL_2019[['Team','mean','OLrank']]
OL_2019
```
